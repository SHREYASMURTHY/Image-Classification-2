```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop(['PassengerId','SibSp','Parch','Ticket','Name','Cabin','Embarked'],axis=1,inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
df['Sex_Encoded']=le_sex.fit_transform(df['Sex'])
df.drop(['Sex'],axis=1,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
x,y=df.drop('Survived',axis=1),df['Survived']
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=56)
```

```python
from sklearn import tree
dt=tree.DecisionTreeClassifier()
dt.fit(train_x,train_y)
```

```python
dt.score(test_x,test_y)
```
