```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
df=load_iris()
df
```

```python
dir(df)
```

```python
df.target_names
```

```python
iris=pd.DataFrame(df.data,columns=df.feature_names)
iris
```

```python
iris
```

```python
iris.drop(['sepal length (cm)', 'sepal width (cm)'], axis=1, inplace=True)
iris
```

```python
iris
```

```python
iris['flower']=df['target']
iris
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
yp=km.fit_predict(iris)
yp
# n_clusters =3 beacuse  you have 3 targets so u create km.km will fit iris dataset 
#and will create 3 clusters in which each clusters will be similar.then km will try to
#predict the yp values bases on the iris dataset
```

```python
iris['yp']=yp
iris
```

```python
inertia = km.inertia_
print(inertia)
```

```python
df1=iris[iris['yp']==0]
df2=iris[iris['yp']==1]
df3=iris[iris['yp']==2]
df1
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='green')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='red')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='black')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
#you took all the similar target values and formed three separate datasets(df1,df2,df3)
#thn you plotted petal length vs petal width for all 3 of them wrt themselves.
```

```python
#red custer ios fine but the black and green cluster is overlapped this is cuz the scaling 
#is not right so we use minmaxscaler 
#we scale attributes so that the values lie in same range
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(iris[['petal length (cm)']])
iris[['petal length (cm)']]=scaler.transform(iris[['petal length (cm)']])
iris
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(iris[['petal width (cm)']])
iris[['petal width (cm)']]=scaler.transform(iris[['petal width (cm)']])
iris
```

```python
#now since we have scaled them we will apply kmeans algo once again and form 3 clsuters
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3) 
yp=km.fit_predict(iris)
yp
iris['yp']=yp
```

```python
#now since the attributes have been scaled we will again create 3 separate dataframes
#and plot them versus their own attributes
df4=iris[iris['yp']==0]
df5=iris[iris['yp']==1]
df6=iris[iris['yp']==2]
plt.scatter(df4['petal length (cm)'],df4['petal width (cm)'],color='green')
plt.scatter(df5['petal length (cm)'],df5['petal width (cm)'],color='red')
plt.scatter(df6['petal length (cm)'],df6['petal width (cm)'],color='black')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
```

```python
#elbow plot
from sklearn.cluster import KMeans
sse=[]
k_range=(1,10)
for i in k_range:
    km=KMeans(n_clusters=i)
    km.fit(iris)
    sse.append(km.inertia_)
    
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_range,sse)
```
