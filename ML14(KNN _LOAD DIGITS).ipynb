```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
df=load_digits()
df
```

```python
dir(df)
```

```python
df.data
```

```python
digits=pd.DataFrame(df.data,columns=df.feature_names)
digits['target']=df.target
digits
```

```python
from sklearn.model_selection import train_test_split
x=df['data']
y=df['target']
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=56)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=5)
kn.fit(train_x,train_y)
```

```python
kn.score(test_x,test_y)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=kn.predict(test_x)
cm=confusion_matrix(test_y,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 12, "color": "black"})
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
